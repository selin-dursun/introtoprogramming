```python
# creates a class called beverage
class Beverage:
# constructor that defines attributes of size and price 
    def __init__(self, size, price): 
        self.size = size
        self.price = price 
        
#method that displays the object, sharing the size and price 
    def display(self):
        return "Size: "+ str(self.size)+ "   Price: ${:.2f}".format(self.price,2)

#creates a SubClass called Soda... it will inherit all the attributes and methods of Beverage
class Soda(Beverage):
    def __init__(self, size, price): 
        Beverage.__init__(self, size, price)
   

#creates a subclass called Lemonade from Beverage
class Lemonade(Beverage):
#the constructor contains all the elements of beverage, but also contains an additional attribute called sugar. 
    def __init__(self, size, price, sugar): 
        Beverage.__init__(self, size, price)
        self.sugar = sugar

#the display function is overriden to include sugar 
    def display(self):
        return Beverage.display(self)+"    Sugar: "+str(self.sugar)

# a function (outside of any classes) is created that displays the title 
def title():
    print("The Beverage Program")

# a function is created prints a closing statements along with the number of drinks created 
def closing(drinks):
        print("\nYou created "+str(len(drinks))+" drink orders. Thanks for using this program!")

# a function called get_size that captures information on what size the drink object is and returns a description of the size.  
def get_size():
    size_input = input("What size would you like?  Enter S, M or L: ")
    if size_input == "M": size = "Medium" 
    elif size_input == "L": size = "Large"
    else: size = "Small" 
    return size

# a function called get_sugar that captures information on whether the lemonade is sweetened or unsweetened 
def get_sugar(): 
    sugar_input = input("Sweetened or Unsweetened?  Enter S or U: ")
    if sugar_input == "S": sugar = "Sweetened"
    else: sugar = "Unsweetened"
    return sugar

# a function that returns a price based on the size 
def get_price(size):
    if size == "Large": price = 2.00
    elif size == "Medium": price = 1.75
    else: price = 1.50
    return price 

### NEED TO ADD THIS UFNCTION TO PROMPT FOR SIMPLIFICATION 
def get_type():
    drink_type = input("\nSoda or a lemonade? Enter S, L or 0 for no drinks: ")
        #getting the user input on the size and converting it to a full string, e.g. Small for Small
    if drink_type !="S" and drink_type !="L" and drink_type !="0": 
        drink_type = ("Incorrect drink type.  Please enter S or L: ")
    return drink_type     

def main():
    title()
    #an empty lists called drinks is created so that the various drink objects created can be added to this list. 
    drinks = []
    again = "y"
    # a loop that runs as long as the user wants to create a drink 
    while again == "y":
        try: 
            drink_type = get_type()
            size = get_size()
            price = get_price(size)
            if drink_type == "S":
                drink = Soda(size, price)
            elif drink_type == "L":
                sugar = get_sugar()
                drink = Lemonade(size, price, sugar)
            elif drink_type == "0": break
            drinks.append(drink)
            again = input("Would you like to create another drink? y/n? ")
        except ValueError: 
            again = input("Incorrect value entered. Create another drink? ")
        except: 
            again = input("Error.  Create another drink? ")
            
    for drink in drinks:
        if isinstance(drink, Lemonade):
            print("\nDrink "+str(drinks.index(drink)+1)+": Lemonade")
        elif isinstance(drink, Soda): 
            print("\nDrink "+str(drinks.index(drink)+1)+": Soda")
        print(drink.display())
    closing(drinks)        
    
if __name__ == "__main__":
     main()
```
