```python
# Campus Pizza Ordering Program
# opens up regular toppings file and saves it as a list called "regular" 
with open ("regular.txt","r") as r:
    regular = r.read().splitlines()
    
# opens up premium toppings file and saves it as a list called "premium"  
with open ("premium.txt","r") as p:
    premium = p.read().splitlines()

def opening():
    print("Campus Pizza Ordering Program") 

def crust_type(crust): 
    if str.lower(crust) == "w": crust_name = "Wheat"
    else: crust_name = "Regular"
    return crust_name 

# this function first takes the list of regular toppings stored in the list "regular" and prints it along with the number (i+1, so that the first topping is a 1 not a 0).  
# The function also allows the user to enter a number, and based ont he number it appends that topping to r_selected, an empty list created in the main funciton. 
def r_selection(regular, r_selected):
    print("\nRegular toppings available at 75 cents each: ")
    #for each element in the list regular, the element is assigned to the variable "topping", and a counter is created where the index of that topping (or the number position it's in) is printed.  
    for i in regular:
        topping = i
        print(str(regular.index(i)+1)+". "+topping)
    reg = "y" 
    while reg != 0: 
        reg = int(input("\nEnter the number of the regular topping, or \"0\" to exit: "))
        if reg == 0: break 
        #This function is used to pull the appropriate element out of the list regular based on the number the user enters, which is stored as "reg". 
        r_new = regular[reg-1]
        # This confirms to the user that the correct topping was added
        print(r_new + " added.") 
        # The topping selected by the user is added to the list r_selected 
        r_selected.append(r_new) 
    # We need to return r_selected as a variable from r_selection 
    return r_selected

#similar function as r_selection but for premium toppings 
def p_selection(premium, p_selected): 
    print("\nPremium toppings available at $1.25 each: ")
    for i in premium:
        topping = i
        print(str(premium.index(i)+1)+". "+topping)
    prem = "y" 
    while prem != 0: 
        prem = int(input("\nEnter the number of the premium topping, or \"0\" to exit: "))
        if prem == 0: break 
        p_new = premium[prem-1]
        print(p_new + " added.") 
        p_selected.append(p_new)
    return p_selected

#calculates a final price based on the number o premium and regular toppings selected by using the len function, which tells you the number of elements in each list (p_selected or r_selected). 
def final(crust, r_selected, p_selected):
    if str.lower(crust) == "w":
        final_price = 8 + (0.75 * float(len(r_selected))) + (1.25 * float(len(p_selected)))
    else: final_price = 7 + (0.75 * float(len(r_selected))) + (1.25 * float(len(p_selected)))
    return final_price

#This statement prints the pizza number (the number comes from the loop in the main function), the crust type,  the list of premium & regular toppings and the price. 
def pizza(number, crust_name, r_selected, p_selected, final_price): 
    print("\nPizza "+str(number))
    print("Crust: "+crust_name)
    print("Regular Toppings:", end=" ")
    for i in r_selected: 
        print(i, end="  ")
    print("\nPremium Toppings:", end=" ")
    for i in p_selected:
        print(i, end="  ")
    print("\nThe final price of the pizza is ${:.2f}".format(final_price)+".")   
    
def closing(number):
    print("\nThanks for using this program! You created "+str(number)+ " pie(s).") 

def main(): 
    opening()
    #set the number of pizza equal to 0, and then increase it each time in the loop to keep track of the pizza #  
    number = 0
    again = "y"
    # set again to yes, and begin a loop that will continue if the user wants to build another pizza 
    while again == "y":
        #the program should be nested inside a try statement so if the user makes an error, we can let the user know 
        try:
            #define r_selected & p_selected as lists 
            r_selected = []
            p_selected = []
            #gather crust input
            crust = input("\nWhat crust (please enter \"w\" for wheat at $8 or \"r\" for regular at $7): ")                      
            crust_name = crust_type(crust)
            #run the r_selection and p_selection function with the inputs of the appropriate list of toppings, and the newly created list that will gather the user selections 
            r_selected = r_selection(regular, r_selected)
            p_selected = p_selection(premium,p_selected)
            #prints the final price of the pizza after the user has selected the toppings 
            final_price = final(crust, r_selected, p_selected)
            #a counter to keep track of the pizza numbers 
            number += 1
            #returns the pizza number, the crust type selected, the toppings selected & final price 
            pizza(number, crust_name, r_selected, p_selected, final_price)
            again = input("\nWould you like to create another pie? y or n? ")           
        except ValueError: print("You entered an incorrect value.  Try again.")
        except: print("You made an error.  Try again.")
    closing(number)
    
if __name__ == "__main__":
     main()
```
